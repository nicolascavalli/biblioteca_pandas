```python
import pandas as pd
import seaborn as sns
```

```python
df_vendas = pd.read_excel('varejo.xlsx')
```

```python
df_vendas.head()
```

```python
df_vendas.info()
```

```python
df_vendas['idcanalvenda'].value_counts()
```

```python
df_vendas['idcanalvenda'] = df_vendas['idcanalvenda'].str.replace('APP','Aplicativo')
```

```python
df_vendas['idcanalvenda'].value_counts()
```

```python
df_vendas['bandeira'].value_counts()
```

```python
df_vendas['Nome_Departamento'].value_counts()
```

```python
df_vendas['estado'].value_counts()
```

```python
df_vendas['Nome_Departamento'] = df_vendas['Nome_Departamento'].str.replace(' ','_')
```

```python
df_vendas['Nome_Departamento'].value_counts()
```

```python
df_vendas.info()
```

```python
df_vendas['estado'].isnull()
```

```python
df_vendas_snulos = df_vendas.dropna()
```

```python
df_vendas_snulos.info()
```

```python
df_vendas['estado'] = df_vendas['estado'].fillna('MS')
```

```python
df_vendas.info()
```

```python
media_preco = df_vendas['Preço'].mean()
```

```python
df_vendas['Preço'].fillna(media_preco, inplace=True)
```

```python
df_vendas.info()
```

```python
df_vendas.describe()
```

```python
df_vendas.info()
```

```python
df_vendas = df_vendas.query('Preço < Preço_com_frete')
```

```python
df_vendas.head()
```

```python
df_vendas.describe()
```

```python
df_vendas.groupby('Nome_Departamento').idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
agg_departamento = round(df_vendas.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending=False).reset_index().head(5),2)
```

```python
df_vendas.groupby('Data').idcompra.nunique().sort_values(ascending=False)
```

```python
df_vendas['mês'] = df_vendas['Data'].dt.month
```

```python
df_vendas.head()
```

```python
agg_data = df_vendas.groupby('mês')['idcompra'].nunique().reset_index()
agg_data
```

```python
df_cliente = pd.read_excel('cliente_varejo.xlsx')
```

```python
df_cliente.head()
```

```python
df_vendas.head()
```

```python
df_cliente.describe()
```

```python
df_cliente.info()
```

```python
df_cliente = df_cliente.astype({'renda':'float64'})
```

```python
df_cliente.info()
```

```python
df_consolidado = df_vendas.merge(df_cliente, how='left', on='cliente_Log')
```

```python
df_consolidado.head()
```

```python
agg_canalvenda = round(df_consolidado.groupby('idcanalvenda')['renda'].agg('mean').sort_values(ascending=False).reset_index(), 2)
```

```python
agg_bandeira = round(df_consolidado.groupby('bandeira')['idade'].agg('mean').sort_values(ascending=False).reset_index(), 2)
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(7,4))
plt.bar(agg_bandeira['bandeira'], agg_bandeira['idade'], color='green')
plt.ylabel('Média de Idade')
plt.title('Idade Média por Bandeira', loc='center', fontsize=18, fontweight='bold', color='black');
```

```python
fig = plt.figure(figsize=(7,4))
plt.bar(agg_canalvenda['idcanalvenda'], agg_canalvenda['renda'], color='blue')
plt.ylabel('Média de Renda')
plt.title('Renda Média por Canal de Venda', loc='center', fontsize=18, fontweight='bold', color='black');
```

```python
fig = plt.figure(figsize=(7,4))
plt.xlabel('Mês')
plt.ylabel('Quantidade de Vendas')
plt.title('Quantidade de Vendas por Mês', loc='center', fontsize=18, fontweight='bold', color='black')
plt.plot(agg_data['mês'], agg_data['idcompra'], color='red', linewidth = 3)
```

```python
fig = plt.figure(figsize=(7,4))
plt.xlabel('Departamento')
plt.ylabel('Média de Preço')
plt.title('Média de Preços por Departamento', loc='center', fontsize=18, fontweight='bold', color='black')
plt.bar(agg_departamento['Nome_Departamento'], agg_departamento['Preço_com_frete'], color='yellow')
```
